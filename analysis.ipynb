# VISUALIZATIONS 


```python
plt.figure(figsize=(12, 9))
sns.set_style("whitegrid")

#Gender Wise Survival Rate 
plt.subplot(2, 2, 1)

sns.barplot(x=survival_rate_by_gender.index , 
            y=survival_rate_by_gender.values , 
            palette='bright' , 
            hue=survival_rate_by_gender.index )

plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.title("Survival Rate VS Gender", fontsize= 12)



#Survival Rate VS Cancer Stage
plt.subplot(2,2,2)
colours = sns.color_palette('coolwarm' , len(Survival_Rate_VS_Cancer_Stage))
sns.barplot(x=Survival_Rate_VS_Cancer_Stage.index , 
            y= Survival_Rate_VS_Cancer_Stage.values , 
            palette=colours , 
            hue= Survival_Rate_VS_Cancer_Stage.index )

plt.ylabel("Survival Rate")
plt.xlabel("Cancer Stage")
plt.title("Survival Rate VS Cancer Stage" , fontsize= 12)



#Survival Rate VS Smoking Habit
plt.subplot(2,2,3)
labels = sns.barplot(x=smoker_survival_percentage.index , y= smoker_survival_percentage.values , palette='muted' ,hue=smoker_survival_percentage.index)

for i in labels.patches :
    labels.annotate(f'{i.get_height() : 0.1f}%' , 
                    (i.get_x() + i.get_width() / 2 , i.get_height() ) , 
                    ha='center' , va= "bottom")

plt.ylabel("Survival Rate")
plt.xlabel("Type Of Smoking Habit")
plt.title("Survival Rate VS Smoking Habit" , fontsize= 12)



#Treatment Wise Survival Rate 
plt.subplot(2,2,4)
sns.barplot(x=treatment_type.index , y = treatment_type.values , palette='dark' ,  hue=treatment_type.index)
plt.ylabel("Survival Rate")
plt.xlabel("Treatment Type")
plt.title("Treatment Wise Survival Rate" , fontsize= 12)

plt.tight_layout(pad=3.0)
plt.show()
```

```python
# Age Distribution Among Cancer Patients

plt.figure(figsize=(10,8))
sns.histplot(df['age'] , bins=30 , kde=True)
plt.title("Age Distribution Among Cancer Patients" , fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Age Categories 

def age_categories(age):
    if age <=30:
        return 'Young'
    elif age>30 and age<=50:
        return 'Middle Aged'
    else:
        return "Old"
df['age_group'] = df['age'].apply(age_categories)

survival_rate_age_group = df.groupby('age_group')['survived'].mean().sort_values(ascending=False )
print(survival_rate_age_group)

#Count of the categories
df['age_group'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=survival_rate_age_group.index , y=survival_rate_age_group.values , palette='bright')
plt.title('Survival rate by age group', fontsize = 11)
plt.xlabel('Age group')
plt.ylabel('Survival rate')
plt.grid(True)
plt.show()
```

```python
survival_treatment_age = df.groupby(['age_group', 'treatment_type'])['survived'].mean().unstack()
survival_treatment_age
```

```python
survival_treatment_age.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Survival Rate by Treatment Type and Age Group', fontsize=12)
plt.xlabel('Age Group', fontsize=8)
plt.ylabel('Survival Rate', fontsize=8)
plt.ylim(0, 1)
plt.legend(title='Treatment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df['diagnosis_date'] = pd.to_datetime(df['diagnosis_date'])
df['end_treatment_date'] = pd.to_datetime(df['end_treatment_date'])
```

```python
df.head()
```

```python
df['date_diff'] = df['end_treatment_date']- df['diagnosis_date']

df.head()
```

```python
result_df = df.groupby(['country', 'cancer_stage']).agg({
    'date_diff': lambda x: x.mean().days,
    'survived': lambda x: round(x.mean() * 100, 2),
    'hypertension': lambda x: round(x.mean() * 100, 2),
    'asthma': lambda x: round(x.mean() * 100, 2),
    'smoking_status': lambda x: x.mode().iloc[0] if not x.mode().empty else None
}).reset_index()


result_df.rename(columns={
    'date_diff': 'avg_treatment_days',
    'survived': 'survival_rate(%)',
    'hypertension': 'hypertension_rate(%)',
    'asthma': 'asthma_rate(%)'
} , inplace=True)

result_df
```

```python
plt.figure(figsize=(18,12))
sns.barplot( data=result_df , x='country' , y='avg_treatment_days', hue='cancer_stage')
plt.title("Average Treatment Duration by Country and Cancer Stage")
plt.ylabel('Average Days')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
